# Stemming

```python
import nltk
from nltk.tokenize import word_tokenize
import string
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
from nltk.stem import PorterStemmer

data = 'Experience the vibrants math maths mathed cleaning culture and scenic beauty of India. Plan your trip with Incredible India & explore tourist destinations, cultural events'
data = data.lower()
data = data.translate(str.maketrans("","",string.punctuation))
print(data,"\n")
data_tokens = word_tokenize(data)
print(data_tokens,"\n")
stemmer = PorterStemmer()
cleaned_tokens = []
for token in data_tokens:
    if token not in stop_words:
        token = stemmer.stem(token)
        cleaned_tokens.append(token)

print(cleaned_tokens,"\n")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk.tokenize import word_tokenize
import string
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import pandas as pd
def clean_text(data):
    data = str(data).lower()
    data = data.translate(str.maketrans("","",string.punctuation))
    data_tokens = word_tokenize(data)
    lemmatizer = WordNetLemmatizer()
    cleaned_tokens = []
    for token in data_tokens:
        if (token not in stop_words) and (token.isalpha()):
            token = lemmatizer.lemmatize(token)
            cleaned_tokens.append(token)
    if len(cleaned_tokens)<5:
        return "NULL"
    else:
        return " ".join(cleaned_tokens)
df = pd.read_csv('../Data/Combined Data.csv')
df = df.head(100)
tokens = []
data = df['statement'].values
for text in data:
    cleaned_tokens = clean_text(text)
    if cleaned_tokens != "NULL":
        tokens.append(cleaned_tokens)
        
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(tokens)
TF_df = pd.DataFrame(vectors.toarray(),columns = vectorizer.get_feature_names_out())
TF_df
```
